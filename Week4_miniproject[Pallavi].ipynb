## Part 1: Enhanced Expense Entry (15 minutes)

### 🔄 From One to Many

Transform your Week 2 single-expense tracker to handle multiple expenses with proper validation:

```python
print("=== Enhanced Finance Tracker v0.3 ===")
print("Week 3: Loop-Powered Edition")
print()

# User setup (enhanced from Week 2)
name = input("Enter your name: ")

# Add budget validation using while loop
monthly_budget = 0
while monthly_budget < 100 or monthly_budget > 10000:
    try:
        budget_input = input("Enter monthly budget ($100-$10,000): $")
        monthly_budget = float(budget_input)

        if monthly_budget < 100:
            print("Budget too low! Minimum $100")
        elif monthly_budget > 10000:
            print("Budget seems high! Maximum $10,000")
    except ValueError:
        print("Please enter a valid number!")

print(f"\nWelcome {name}! Monthly budget: ${monthly_budget:.2f}")

# ========== MULTI-EXPENSE ENTRY ==========
expenses = []  # Store all expense amounts
descriptions = []  # Store what each expense was for
total_spent = 0

print("\n--- Enter Your Expenses ---")
print("Type 'done' when finished entering expenses")

# Your task: Complete the multi-expense entry system

# Use sentinel value pattern to collect multiple expenses
while True:
    expense_input = input("\nExpense amount (or 'done'): $")

    if expense_input.lower() == 'done':
        break

    # Validate each expense amount (positive numbers only)
    try:
        amount = float(expense_input)
        if amount <= 0:
            print("Expense amount must be positive!")
            continue # Ask for amount again if not positive
    except ValueError:
        print("Please enter a valid number!")
        continue # Ask for amount again if not a number

    # Store both amount and description
    description = input("What was this expense for?: ")
    expenses.append(amount)
    descriptions.append(description)

    # Show running total and budget percentage after each entry
    total_spent += amount
    print(f"Total spent: ${total_spent:.2f}")
    budget_percentage = (total_spent / monthly_budget) * 100
    print(f"Budget used: {budget_percentage:.1f}%")

    if total_spent > monthly_budget:
        print("You've gone over budget!")


print(f"\n📊 Expense entry complete: {len(expenses)} expenses, ${total_spent:.2f} total")
```

## Part 2: Basic Expense Analysis (15 minutes)

### 📈 Processing Your Transaction Data

Use loops to analyze the expenses you collected:

```python
# Basic statistics using loop patterns from workshops
# Added total_spent calculation to ensure variable exists
total_spent = sum(expenses)
print(f"Total expenses: {len(expenses)}")
print(f"Total spent: ${total_spent:.2f}")
print(f"Average expense: ${total_spent/len(expenses):.2f}")
print(f"Budget used: {(total_spent/monthly_budget)*100:.1f}%")
print(f"Remaining budget: ${monthly_budget - total_spent:.2f}")

# Your task: Find highest and lowest expenses
# Use the min/max pattern from workshops
highest_expense = 0
lowest_expense = 0

# Your min/max finding code here
highest_expense = expenses[0]
lowest_expense = expenses[0]

# Added loop to find actual highest and lowest
for expense in expenses:
    if expense > highest_expense:
        highest_expense = expense
    if expense < lowest_expense:
        lowest_expense = expense

print(f"\n💰 Expense Range:")
print(f"Highest expense: ${highest_expense:.2f}")
print(f"Lowest expense: ${lowest_expense:.2f}")

# Your task: Categorize expenses by size
# Count how many expenses fall into each category:
# Large (>20% of budget), Medium (5-20%), Small (<5%)

large_count = 0    # > 20% of budget
medium_count = 0   # 5-20% of budget
small_count = 0    # < 5% of budget

# Your categorization loop here
for expense in expenses:
  percent = (expense / monthly_budget) * 100
  if percent > 20:
    large_count += 1
  elif percent >= 5 and percent <= 20:
    medium_count += 1
  else:
    small_count += 1

print(f"\n📊 Expense Categories:")
print(f"Large expenses (>20% budget): {large_count}")
print(f"Medium expenses (5-20% budget): {medium_count}")
print(f"Small expenses (<5% budget): {small_count}")

# Your task: Provide smart feedback based on the analysis
# Give different advice based on spending patterns
print("\n💡 Smart Feedback:")

# Your feedback logic here
# Consider: budget percentage used, number of large expenses, etc.
total_percent = (total_spent / monthly_budget) * 100
if total_percent > 100:
  print("You've gone over budget!")
  print("Advice: Review your expenses")
elif total_percent > 80:
  print("You're spending a lot of money!")
  print("Advice: Consider reducing expenses")
elif total_percent >20:
  print("You're spending a decent amount of money!")
  print("Advice: Consider savings also")
else:
  print("You're spending very little relative to your budget!")
  print("Advice: Great job! Consider moving funds to savings.")
```

## Part 3: Multiple Savings Goals Tracker (15 minutes)

### 🎯 Goal Management with Loops

Extend your tracker to handle multiple savings goals:

```python
print("\n=== Savings Goals Tracker ===")

# Goal collection
goal_names = []
goal_targets = []
goal_current = []

# Ask how many goals they want to track
num_goals = 0
while num_goals < 1 or num_goals > 5:  # Reasonable limits
    try:
        num_goals = int(input("How many savings goals? (1-5): "))
        if num_goals < 1:
            print("Must have at least 1 goal!")
        elif num_goals > 5:
            print("Maximum 5 goals for this tracker!")
    except ValueError:
        print("Please enter a valid number!")

# Your task: Collect goal information using for loop
# For each goal (1 to num_goals):
#   - Get goal name
#   - Get target amount (with validation)
#   - Get current amount (with validation)
#   - Store in the three lists

print(f"\nEntering {num_goals} savings goals:")

for i in range(num_goals):
    print(f"\n--- Goal {i+1} ---")

    # Your goal collection code here
    name = input("Enter goal name: ")
    goal_names.append(name)

    target_amount = 0
    while target_amount <= 0:
        try:
            target_amount = float(input(f"Enter target amount for {name}: $"))
            if target_amount <= 0:
                print("Target amount must be positive!")
        except ValueError:
            print("Please enter a valid number!")
    goal_targets.append(target_amount)

    current_amount = -1 # Initialize with a value that ensures the loop runs
    while current_amount < 0 or current_amount > target_amount:
        try:
            current_amount = float(input(f"Enter current saved amount for {name}: $"))
            if current_amount < 0:
                print("Current amount cannot be negative!")
            elif current_amount > target_amount:
                print(f"Current amount (${current_amount:.2f}) cannot exceed target amount (${target_amount:.2f})!")
        except ValueError:
            print("Please enter a valid number!")
    goal_current.append(current_amount)


print(f"\n✅ Saved {len(goal_names)} goals!")
```

```python
# Goal Analysis and Progress Tracking
if len(goal_names) > 0:
    print("\n📊 Goals Progress Analysis:")

    # Your task: Analyze each goal
    # For each goal, calculate:
    #   - Progress percentage
    #   - Amount remaining
    #   - Months to completion (if saving $100/month)

    total_target = 0
    total_current = 0

    lowest_progress_percentage = 101.0 # Initialize with a value higher than 100%
    goal_needing_most_attention_index = -1

    for i in range(len(goal_names)):
        # Your progress calculation here
        name = goal_names[i]
        target = goal_targets[i]
        current = goal_current[i]

        progress_percentage = (current / target) * 100 if target > 0 else 0
        amount_remaining = target - current

        months_to_completion = "N/A"
        if amount_remaining > 0:
            months_to_completion = amount_remaining / 100 # Assuming $100/month saving
            months_to_completion = f"{months_to_completion:.1f} months"
        elif amount_remaining <= 0:
            months_to_completion = "Achieved!"

        print(f"\n--- {name} ---")
        print(f"Target: ${target:.2f}")
        print(f"Current: ${current:.2f}")
        print(f"Progress: {progress_percentage:.1f}%")
        print(f"Remaining: ${amount_remaining:.2f}")
        print(f"Months to complete (at $100/month): {months_to_completion}")

        total_target += target
        total_current += current

        if progress_percentage < lowest_progress_percentage:
            lowest_progress_percentage = progress_percentage
            goal_needing_most_attention_index = i

    # Overall progress summary
    overall_progress = (total_current / total_target) * 100 if total_target > 0 else 0
    total_remaining = total_target - total_current

    print(f"\n🎯 Overall Goals Summary:")
    print(f"Total target: ${total_target:.2f}")
    print(f"Total saved: ${total_current:.2f}")
    print(f"Overall progress: {overall_progress:.1f}%")
    print(f"Total remaining: ${total_remaining:.2f}")

    # Your task: Find which goal needs most attention
    # Find the goal with lowest progress percentage
    # Provide encouragement and specific advice
    if goal_needing_most_attention_index != -1:
        attention_goal_name = goal_names[goal_needing_most_attention_index]
        print(f"\n💡 Goal needing most attention: {attention_goal_name}")
        print(f"You're at {lowest_progress_percentage:.1f}% progress for {attention_goal_name}. Keep going!")
        if goal_targets[goal_needing_most_attention_index] - goal_current[goal_needing_most_attention_index] > 0:
            print(f"Consider setting aside an extra amount each month for this goal to reach it faster.")
        else:
            print(f"Great job! You've already reached or surpassed this goal. Consider a new target!")

else:
    print("No goals to analyze!")
```

## Part 4: Simple Menu System (10 minutes)

### 🎮 Interactive Program Flow

Create a basic menu that lets users choose what to do:

```python
print("\n=== Interactive Finance Menu ===")

# Simple menu system
running = True

while running:
    print("\n" + "-"*30)
    print(f"💰 {name}'s Finance Tracker")
    if len(expenses) > 0:
        print(f"📊 Current: {len(expenses)} expenses, ${sum(expenses):.2f} spent")
    print("-"*30)
    print("1. View expense summary")
    print("2. View goals progress")
    print("3. Add new expense")
    print("4. Budget check")
    print("5. Exit")

    choice = input("\nChoose option (1-5): ")

    if choice == "1":
        # Display expense summary
        print("\n--- Expense Summary ---")

        # Your task: Show all expenses with numbers
        # Format: "1. $23.50 - Coffee"
        # Include total at the end
        if expenses:
            for i in range(len(expenses)):
                print(f"{i+1}. ${expenses[i]:.2f} - {descriptions[i]}")
            print(f"Total spent: ${sum(expenses):.2f}")
        else:
            print("No expenses recorded yet.")


    elif choice == "2":
        # Display goals progress
        print("\n--- Goals Progress ---")

        # Your task: Show each goal with progress percentage
        # Format: "Goal Name: $current/$target (XX% complete)"
        if goal_names:
            for i in range(len(goal_names)):
                name = goal_names[i]
                target = goal_targets[i]
                current = goal_current[i]
                progress = (current / target) * 100 if target > 0 else 0
                print(f"{name}: ${current:.2f}/${target:.2f} ({progress:.1f}% complete)")
        else:
            print("No savings goals set yet.")


    elif choice == "3":
        # Add single new expense
        print("\n--- Add New Expense ---")

        # Your task: Add one more expense to existing lists
        # Use validation pattern, update totals, give feedback
        while True:
            expense_input = input("Enter new expense amount: $")
            if expense_input.lower() == 'done':
                break # Allow 'done' to exit adding single expense if user changes mind
            try:
                amount = float(expense_input)
                if amount <= 0:
                    print("Expense amount must be positive!")
                    continue
                break
            except ValueError:
                print("Invalid amount! Please enter a number.")

        if expense_input.lower() != 'done': # Only proceed if user didn't type 'done'
            description = input("What was this new expense for?: ")
            expenses.append(amount)
            descriptions.append(description)
            total_spent += amount # Update total_spent immediately
            print(f"Added: ${amount:.2f} for {description}")
            print(f"Current total spent: ${total_spent:.2f}")
            if total_spent > monthly_budget:
                print("Warning: You are now over budget!")
            else:
                print(f"Remaining budget: ${monthly_budget - total_spent:.2f}")

    elif choice == "4":
        # Budget status check
        print("\n--- Budget Check ---")

        # Your task: Calculate and display budget status
        # Show percentage used, remaining amount, daily budget left
        # Give advice based on budget usage
        if monthly_budget > 0:
            current_total_spent = sum(expenses) # Recalculate to be sure, or use existing total_spent
            percent_used = (current_total_spent / monthly_budget) * 100
            remaining_budget_amount = monthly_budget - current_total_spent

            print(f"Monthly Budget: ${monthly_budget:.2f}")
            print(f"Total Spent: ${current_total_spent:.2f}")
            print(f"Budget Used: {percent_used:.1f}%")
            print(f"Remaining Budget: ${remaining_budget_amount:.2f}")

            if remaining_budget_amount > 0:
                # Simple estimate for daily budget remaining - assuming ~30 days in a month
                days_in_month = 30
                print(f"Considered daily average for remaining budget: ${remaining_budget_amount / days_in_month:.2f}/day")

            if percent_used > 100:
                print("Advice: You are over budget! Time to review and cut down on spending.")
            elif percent_used > 80:
                print("Advice: You're getting close to your budget limit. Spend carefully!")
            elif percent_used < 50:
                print("Advice: Great job managing your budget! You have plenty of room.")
            else:
                print("Advice: You're on track with your budget. Keep it up!")
        else:
            print("Monthly budget not set or is zero. Cannot perform budget check.")

    elif choice == "5":
        # Exit with summary
        running = False
        print(f"\nGoodbye {name}!")
        print(f"Final summary: {len(expenses)} expenses, ${sum(expenses) if expenses else 0:.2f} total")
        print("🚀 Your finance tracker is growing stronger each week!")

    else:
        print("Invalid choice! Please choose 1-5.")

print("\n=== Finance Tracker Session Complete ===")
```

## Part 5: Testing Your Enhanced Tracker (10 minutes)

### 🧪 Complete System Test

Test your loop-powered finance tracker with realistic data:

```python
# Test your tracker with this scenario:
test_instructions = """
🧪 TEST SCENARIO:

Setup:
- Name: enter name
- Budget: $2000

Goals (try 2 goals):
- "Emergency Fund": target $1000, current $250
- "Vacation": target $500, current $50

Expenses to enter:
- $450 for "Rent payment"
- $67.50 for "Weekly groceries"
- $89.00 for "Gas for car"
- done

Then test each menu option:
1. View expense summary (should show 3 expenses)
2. View goals progress (should show percentages)
3. Add expense: $25 for "Coffee"
4. Budget check (should show updated totals)
5. Exit

Expected final result:
- 4 expenses totaling $631.50
- 31.6% of budget used
- Emergency fund 25% complete
- Vacation fund 10% complete
"""

print(test_instructions)
print("\n⬆️ Run your enhanced tracker above with this test data!")
print("Check that all numbers match the expected results.")


def main():
    print("=== Budget Tracker ===")

    name = input("Enter your name: ")
    budget = float(input("Enter your total budget: $"))

    # Goals
    goals = {}
    for i in range(2):
        print(f"\nEnter goal {i+1}:")
        goal_name = input("Goal name: ")
        target = float(input("Target amount: $"))
        current = float(input("Current saved: $"))
        goals[goal_name] = {"target": target, "current": current}

    # Expenses
    expenses = []

    print("\nEnter your expenses (type 'done' to finish):")
    while True:
        entry = input("Enter expense amount (or 'done'): ")
        if entry.lower() == "done":
            break
        try:
            amount = float(entry)
            if amount <= 0:
                print("Expense amount must be positive!")
                continue
        except ValueError:
            print("Please enter a valid number for the amount!")
            continue
        desc = input("Description: ")
        expenses.append({"amount": amount, "desc": desc})

    # Menu loop
    while True:
        print("\n=== MENU ===")
        print("1. View expense summary")
        print("2. View goals progress")
        print("3. Add expense")
        print("4. Budget check")
        print("5. Exit")

        choice = input("Choose an option: ")

        if choice == "1":
            print("\n--- Expenses ---")
            total = 0
            for e in expenses:
                print(f"${e['amount']:.2f} - {e['desc']}")
                total += e["amount"]
            print(f"Total expenses: ${total:.2f}")

        elif choice == "2":
            print("\n--- Goals Progress ---")
            for goal_name, g in goals.items():  # fixed variable name
                percent = (g["current"] / g["target"]) * 100
                print(f"{goal_name}: {percent:.1f}% complete")

        elif choice == "3":
            while True:
                amount_input = input("Enter amount: $")
                try:
                    amount = float(amount_input)
                    if amount <= 0:
                        print("Amount must be positive!")
                        continue
                    break
                except ValueError:
                    print("Invalid amount! Please enter a number.")

            desc = input("Description: ")
            expenses.append({"amount": amount, "desc": desc})
            print("Expense added.")

        elif choice == "4":
            total = sum(e["amount"] for e in expenses)
            percent_used = (total / budget) * 100
            print(f"\nTotal spent: ${total:.2f}")
            print(f"Budget used: {percent_used:.1f}%")

        elif choice == "5":
            print("Goodbye 👋")
            break

        else:
            print("Invalid option, try again.")


# ✅ THIS IS THE IMPORTANT PART (runs your program)
if __name__ == "__main__":
    main()
```
