```python
print("=== Personal Finance Tracker v0.1 ===")
print("Week 3: Smart Budget Decisions")
print()

# User setup
name = input("Enter your name: ")
monthly_income = float(input("Enter your monthly income: $"))

print(f"\nWelcome to your finance tracker, {name}!")
print(f"Monthly income: ${monthly_income:.2f}")

# This week's feature: Expense categorisation
print("\n--- Expense Entry ---")
expense_amount = float(input("Enter the expense amount: $"))
expense_description = input("What was this expense for? ")

#Category Selection
if "Food" in expense_description:
  category = "Food/Groceries"
  print("This is a food expense")
elif "Transport" in expense_description:
  category = "Transport"
  print("This is a transport expense")
elif "Entertainment" in expense_description:
  category = "Entertainment"
  print("This is an entertainment expense")
elif "Health" in expense_description:
  category = "Health"
  print("This is a health expense")
else:
  category = "Other"
  print("This is an other expense")

#Budget
percentage_of_budget = (expense_amount/monthly_income)*100

print("\n---Expense Summary---")
print(f"Category: {category}")
print(f"Expense Amount: ${expense_amount:.2f}")
print(f"This is {percentage_of_budget:.2f}% of your monthly income.")

#Advice
if percentage_of_budget > 25:
  print("This is a high expense")
  print("Advice: Consider your budget before spending")
elif percentage_of_budget > 10 and percentage_of_budget < 25:
  print("This is a moderate expense")
  print("Advice: Make sure it is on the budget.")
else:
  print("This is a small expense")
  print("Advice: Keep track of your expenses")
```
